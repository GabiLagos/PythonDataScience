EJemplo sobre IRIS, (datos sobre un tipo de flor)

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("data/iris.data.csv")
df
```

Mostramos las últimas filas

```python
df.tail()
```

En este caso no hay cabacera (header), Podemos establecer columnas para el DATAFRAME

```python
df.columns= ["long_sepalo", "ancho_sepalo", "long_petalo", "ancho_petalo", "clase"]
df
```

```python
df.describe()
```

Podemos contar valores en una columna

```python
df["clase"].value_counts()
```

Ordenar datos ascendente o descendente

```python
#df.sort_values('ancho_sepalo')
df.sort_values('ancho_sepalo', ascending=False)
```

Podemos cambiar filas por columnas

```python
df.T
```

Filtrar ppor una columna

```python
df[df["long_petalo"]>=5]
```

Filtrar por mas de una condición.
Se utiliza & para AND 
Se utiliza | para OR

```python
df[
    (df["long_petalo"]>=5)
    &
    (df["ancho_petalo"]>=1.5)
    ]
```

Comprobar valores pedidos dentro del DATAFRAME

```python
df.isna().sum()
```

Vamos aponer valores NaN a alguna columna/fila

```python
df['long_petalo'][:2]=np.nan
df.head()
```

Mostrar valores nulos

```python
df.isna().sum()
```

Podemos quitar los valores si sabemos donde están con fillnan

```python
df['long_petalo']=df['long_petalo'].fillna(2.9)
df.head()
```
